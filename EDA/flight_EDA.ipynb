```python
df = pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.shape
```

```python
df['Date_of_Journey'].str[0:5]
```

```python
print(type(df['Date_of_Journey']))
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]

df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.drop(columns=['Date_of_Journey'],axis=1,inplace=True)
df.head()
```

```python
df['Dep_hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df.drop(columns=['Dep_Time'],axis=1,inplace=True)
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'].mode()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].fillna(df['Total_Stops'].mode()[0])
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.head()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].str.split(' ').str[0]
```

```python
df['Arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.drop(columns=['Arrival_Time'],inplace=True,axis=1)
df.head()
```

```python
df['Additional_Info'].unique()
```

```python
df['Duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]
```

```python
df.head()
```

```python
df['Duration_min'].unique()
```

```python
df['Duration_min'] = df['Duration_min'].fillna(0)
```

```python
df.head()
```

```python
df.drop(columns=['Duration','Route'],inplace=True,axis=1)
```

```python
df.head()
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('googleplaystore.csv')
df.head()
```

```python
df.info()
```

```python
df['Reviews'].unique()
# df['Reviews'].astype(int)
```

```python
df[~df['Reviews'].str.isnumeric()]
```

```python
df_copy = df.copy()
```

```python
# Drop the row with the index 10472
df_copy.drop(index=10472, inplace=True)
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Reviews'] = df_copy['Reviews'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'] = df_copy['Size'].str.replace('M','000')
df_copy['Size'] = df_copy['Size'].str.replace('k','')
df_copy['Size'] = df_copy['Size'].replace('Varies with device',np.nan)
# df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy['Size'] = df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy.head()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
remove_char = ['$',',','+']
columns = ['Price','Installs']
for char in remove_char:
    for col in columns:
        df_copy[col] = df_copy[col].str.replace(char, '')
```

```python
df_copy['Installs'].unique()
df_copy['Installs'] = df_copy['Installs'].astype(int)
```

```python
df_copy['Price'].unique()
df_copy['Price'] = df_copy['Price'].astype(float)
```

```python
df_copy.head()
```

```python
df_copy['Last Updated'] = pd.to_datetime(df_copy['Last Updated'])
df_copy['Day'] = df_copy['Last Updated'].dt.day
df_copy['Month'] = df_copy['Last Updated'].dt.month
df_copy['Year'] = df_copy['Last Updated'].dt.year
df_copy.drop(columns=['Last Updated'],axis=1,inplace=True)
```

```python
df_copy.head()
```

```python
df_copy.info()
```

```python
df_copy.to_csv('googleplay_cleaned_data.csv')
```

```python
df_copy[df_copy.duplicated('App')]
```

```python
df_copy.drop_duplicates(subset=['App'],keep='first',inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Count plot with hue to show distribution across another category
sns.displot(data=df_copy, x="Rating", kde=True)
sns.displot(data=df_copy, x="Size", kde=True)
# Show the plot
plt.show()
```

```python
rating_median = np.median(df_copy['Rating'].dropna())
size_median = np.median(df_copy['Size'].dropna())

df_copy['Rating'].fillna(rating_median, inplace=True)
df_copy['Size'].fillna(size_median, inplace=True)
```

```python
from sklearn.datasets import make_moons
```

```python
x,y = make_moons(n_samples=300,noise=0.06)
```

```python
x
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1],c=y)
```

```python
from sklearn.preprocessing import StandardScaler
sclar = StandardScaler()
```

```python
x_scaled = sclar.fit_transform(x)
x_scaled
```

```python
from sklearn.cluster import DBSCAN
dbscsn = DBSCAN(eps=0.3)
```

```python
dbscsn.fit(x_scaled)
```

```python
dbscsn.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=dbscsn.labels_)
```

```python
from sklearn.datasets import make_circles
x,y = make_circles(n_samples=750,factor=0.3,noise=0.1)
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.cluster import DBSCAN
dbscsn = DBSCAN(eps=0.1)
```

```python
dbscsn.fit_predict(x)
```

```python
plt.scatter(x[:,0],x[:,1],c = dbscsn.labels_)
```

```python
from sklearn.neighbors import LocalOutlierFactor
outlier = LocalOutlierFactor(n_neighbors=4)
```

```python
pred2 = outlier.fit_predict(x)
```

```python
outlier.negative_outlier_factor_
```

```python
outlier.offset_
```

```python
plt.scatter(x[:,0],x[:,1],c = pred2)
```

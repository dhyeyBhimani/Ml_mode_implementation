```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
x,y =make_blobs(n_samples=1000,n_features=2,centers=3,random_state=10)
```

```python
x
```

```python
y
```

```python
plt.scatter(x[:,0],x[:,1],c = y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33,random_state = 10)
```

```python
from sklearn.preprocessing import StandardScaler
scalr = StandardScaler()
```

```python
x_train_scaled = scalr.fit_transform(x_train)
x_test_scaled = scalr.transform(x_test)
```

```python
from sklearn.cluster import KMeans
wss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train_scaled)
    wss.append(kmeans.inertia_)
```

```python
wss
```

```python
plt.plot(range(1,11),wss)
plt.xticks(range(1,11))
plt.xlabel("no of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
kmeans.fit_predict(x_train_scaled)
```

```python
y_pred = kmeans.predict(x_test_scaled)
```

```python
plt.scatter(x_test[:,0],x_test[:,1],c = y_pred)
```

```python
from sklearn.metrics import silhouette_score
```

```python
shil_coef = []
for k in range(2,11):
    kmeans = KMeans(n_clusters=k,init='k-means++')
    kmeans.fit(x_train_scaled)
    score = silhouette_score(x_train_scaled,kmeans.labels_)
    shil_coef.append(score)
    
```

```python
shil_coef # score range is between -1 to 1 and  and more near to +1 better clustering model we have created
```

```python
plt.plot(range(2,11),shil_coef)
plt.xticks(range(1,11))
plt.xlabel("no of clusters")
plt.ylabel("WCSS")
plt.show()
```

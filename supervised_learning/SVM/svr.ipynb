```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
df.info()
```

```python
df['day'].unique()
```

```python
x=df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y=df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.20,random_state = 10)
```

```python
from sklearn.preprocessing import LabelEncoder
l1 = LabelEncoder()
l2 = LabelEncoder()
l3 = LabelEncoder()
```

```python
x_train['sex'] = l1.fit_transform(x_train['sex'])
x_train['smoker'] = l2.fit_transform(x_train['smoker'])
x_train['time'] = l3.fit_transform(x_train['time'])
```

```python
x_train.head()
```

```python
x_test['sex'] = l1.transform(x_test['sex'])
x_test['smoker'] = l2.transform(x_test['smoker'])
x_test['time'] = l3.transform(x_test['time'])
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
one_en = OneHotEncoder()
```

```python
one_en.fit_transform(x_train[['day']]).toarray()
```

```python
df1 = pd.DataFrame(one_en.fit_transform(x_train[['day']]).toarray(),columns=one_en.get_feature_names_out())
```

```python
x_train.drop(columns=['day'],inplace=True,axis=1)
result = pd.concat([df1, x_train], axis=1, join='inner')
display(result)
```

```python
one_en.transform(x_test[['day']]).toarray()
df2 = pd.DataFrame(one_en.transform(x_test[['day']]).toarray(),columns=one_en.get_feature_names_out())
x_test.drop(columns=['day'],inplace=True,axis=1)
result1 = pd.concat([df2, x_test], axis=1, join='inner')
display(result1)
```

```python
from sklearn.svm import SVR
svr = SVR()
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred = svr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
print(r2_score(y_pred,y_test))
print(mean_absolute_error(y_pred,y_test))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100,1000],
              'kernel':['rbf'],
              'gamma':[1,0.1,0.01,0.001]}
```

```python
grid = GridSearchCV(SVR(),param_grid=param_grid,cv = 5,verbose=3,refit=True)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred1 = grid.predict(x_test)
print(r2_score(y_pred1,y_test))
print(mean_absolute_error(y_pred1,y_test))
```

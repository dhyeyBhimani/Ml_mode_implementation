```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
```

```python
print(iris['DESCR'])
```

```python
x = pd.DataFrame(iris['data'],columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'],index=None)
```

```python
x
```

```python
y = iris['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.20,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_dec = DecisionTreeClassifier(max_depth=2)
```

```python
tree_dec.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_dec,filled=True)
```

```python
y_pred = tree_dec.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
print(confusion_matrix(y_pred,y_test))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
df
```

```python
df.head()
```

```python
df.info()
```

Data cleaning

```python
df[df.isnull().any(axis = 1)] #0 for row and 1 for columns
```

```python
df.loc[:122,"region"]  = 0
df.loc[122:,"region"]  = 1
df.info()
```

```python
df[['region']] = df[['region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

removing the null values

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.isnull().sum()
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.info()
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
col = [feature for feature in df.columns if df[feature].dtypes == 'O']
```

```python
col
```

```python
for i in col:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('Algerian_forest_fires_dataset_cleaned.csv')
```
